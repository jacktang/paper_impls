# Mcerp

```python
from mcerp import N, Exp, correlation_matrix, plotcorr
import matplotlib.pyplot as plt

x1 = N(24, 1)
x2 = N(37, 4)
x3 = Exp(2)

x1.describe()
# x1.plot()
# plt.show()
```

```python
print(correlation_matrix([x1, x2, x3]))
plotcorr([x1, x2, x3], labels=['x1', 'x2', 'x3'], full=False)
```

```python
Z = (x1*x2**2)/(15*(1.5 + x3))
Z
```

```python
Z.describe()
Z.plot()
plt.show()
```
